## Exploratory Data Analysis (EDA)

### Explore ...

```python

season_map = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
day_df["season_label"] = day_df["season"].map(season_map)
print(day_df.describe())


```

```python
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
print(hour_df.describe())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="season_label", y="cnt", data=day_df)
plt.title("Rata-rata Peminjaman Sepeda per Musim")
plt.xlabel("Musim")
plt.ylabel("Jumlah Peminjaman")
plt.show()


```

### Pertanyaan 2:

```python
plt.figure(figsize=(10, 5))
hourly_counts = hour_df.groupby("hr")["cnt"].mean().reset_index()
sns.lineplot(x="hr", y="cnt", data=hourly_counts, marker="o", color="b")
plt.title("Rata-rata Peminjaman Sepeda per Jam dalam Sehari")
plt.xlabel("Jam")
plt.ylabel("Jumlah Peminjaman")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```
